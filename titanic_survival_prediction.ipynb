```python
#import the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#load the dataset 
import pandas as pd
df =pd.read_csv("titanic.csv")
#display the first five columns
df.head()
```

```python
#find missing values
df.isnull().sum()
```

```python
## Gives a summary of the dataset including column names, data types, and missing values
df.info()
```

```python
# Displays basic statistics for numeric columns like mean, min, max, and quartiles
df.describe()
```

```python
#fill missing age values in the "age"column with the median age
df["Age"]=df["Age"].fillna(df["Age"].median())
```

```python
# Fill missing values in the 'Embarked' column with the most common port
df["Embarked"]=df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
# Drop the Cabin column and reassign the DataFrame
df=df.drop("Cabin",axis=1)
```

```python
df.info
```

```python
#check for missing values in the "Fare" column
df["Fare"].isnull().sum
```

```python
#check the data types
df.dtypes
```

```python
df["Sex"]=df["Sex"].map({"male" :0 ,"female" : 1})
```

```python
# Convert 'Embarked' column to numeric values: S=0, C=1, Q=2
df["Embarked"]=df["Embarked"].map({"S" : 0,"C": 1,"Q": 2})
```

```python
# Drop columns that are not useful for prediction
df=df.drop(["Name","Ticket","PassengerId"],axis=1)
```

```python
# Check the structure of the cleaned dataset
df.info()
```

```python
# Define features (X) and target (y)
x=df.drop(["Survived"], axis=1)# All columns except 'Survived'
y=df["Survived"]#Target column is 'Survived'
```

```python
# Split the data into training and testing sets (80% train, 20% test)
# Split the data into training and testing sets
# test_size=0.2 means 20% will be used for testing, 80% for training
# random_state=42 helps us get the same results every time
from sklearn.model_selection import train_test_split

# ‚úÇÔ∏è Split the data
X = df.drop("Survived", axis=1)  # Features
y = df["Survived"]               # Target

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# üì¶ Import the model
from sklearn.linear_model import LogisticRegression

# üéì Train the model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
# üì¶ Import accuracy_score
from sklearn.metrics import accuracy_score

# üìä Make predictions
y_pred = model.predict(X_test)

# ‚úÖ Evaluate
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
from sklearn.metrics import confusion_matrix

# üîç Generate confusion matrix
cm = confusion_matrix(y_test, y_pred)

cm
```

```python
from sklearn.metrics import classification_report

# üìã Get precision, recall, F1-score
report = classification_report(y_test, y_pred)

print(report)
```

### üßæ Classification Report
This report shows precision, recall, and F1-score for each class (0 = did not survive, 1 = survived). Higher values = better performance!
